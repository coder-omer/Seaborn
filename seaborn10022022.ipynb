```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
# filter
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
# add new column
cars['type'] =  ['old' if x<=76 else 'new' for x in cars.model_year]
```

```python
cars.head()
```

```python
cars.shape
```

```python
sns.set_style('white')
```

```python
sns.catplot(x='horsepower', data=cars);
```

```python
sns.catplot(y='horsepower', data=cars);
```

```python
sns.catplot(x='horsepower', data=cars, kind='box');
```

```python
sns.catplot(x='horsepower', data=cars, kind='violin');
```

```python
sns.catplot(x='origin', y='horsepower', data=cars, kind='box');
```

```python
sns.catplot(x='origin', y='horsepower', data=cars, hue='cylinders', kind='box');
```

```python
sns.catplot(x='origin', y='horsepower', data=cars, hue='cylinders', kind='violin');
```

```python
sns.set_style('darkgrid')
```

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='box');
```

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='box', col='origin');
```

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='strip', col='origin');
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip', 
            col='origin',
            row='type',
            row_order=['new','old']
            );
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box', 
            col='origin',
            palette='winter',
            whis=1
            );
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip', 
            col='origin',
            s=12,
            alpha=0.3,
            linewidth=2
            );
```

```python
g = sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box', 
            col='origin'
            );
```

```python
sns.set(style='darkgrid', font_scale=1.5)

g = sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box', 
            col='origin'
            );

g.set_titles(col_template="{col_name} cars");
```
